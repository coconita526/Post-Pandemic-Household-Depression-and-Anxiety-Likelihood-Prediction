#### 2.2.1 Identified 4 columns to get the target columns

Since the main goal of original dataset is collecting househould information post-pandemic, there are not target columns of the occurence of anxiety and depression yet. However, among 247 columns there are four columns (refered to four survey questions) specified by National Center for Health Statistics (NCHS), who partnered with Census Bureau in this survey to collect experimental data of household to obtain information on the frequency of anxiety and depression symptoms.

The refered four quesions are as bellow : 

**Adapted PHQ-2 questions**:

Over the last 7 days, how often have you been bothered by … having little interest or pleasure in doing things? Would you say not at all, several days, more than half the days, or nearly every day? Select only one answer.

Over the last 7 days, how often have you been bothered by … feeling down, depressed, or hopeless? Would you say not at all, several days, more than half the days, or nearly every day? Select only one answer.

**Adapted GAD-2 questions**:

Over the last 7 days, how often have you been bothered by the following problems … Feeling nervous, anxious, or on edge? Would you say not at all, several days, more than half the days, or nearly every day? Select only one answer.

Over the last 7 days, how often have you been bothered by the following problems … Not being able to stop or control worrying? Would you say not at all, several days, more than half the days, or nearly every day? Select only one answer.

Reference can be found on : https://www.cdc.gov/nchs/covid19/pulse/mental-health.htm

After checking through the data dictionary, which recorded the survey questions, we know the four features are : 'INTEREST','DOWN','ANXIOUS', and 'WORRY'. 'INTEREST' and 'DOWN' are related to the sympton of 'Major Depressive Disorder' while the 'ANXIOUS' and 'WORRY' are related to the sympton of 'Generalized anxiety disorder'.

#### 2.2.2 Apply the CDC scoring method to get target columns

Fortunately, CDC also published its scoring method to get the diagnose of anxiety and depression based on the above 4 features.
Please read the following reference:

**Scoring and Estimation**

For each scale, the answers are assigned a numerical value: not at all = 0, several days = 1, more than half the days = 2, and nearly every day = 3. The two responses for each scale are added together. A sum equal to three or greater on the PHQ-2 has been shown to be associated with diagnoses of major depressive disorder.  A sum equal to three or greater on the GAD-2 has been shown to be associated with diagnoses of generalized anxiety disorder. For adults with scores of 3 or greater, further evaluation by a clinician or other health professional is generally recommended.

Estimates on this page are based on these composite scores. Answers to both questions in the scale were required to calculate the scores. Adults with missing responses to one or both questions are not shown separately in the tables and are not included in the calculation of percentages.

Reference can be found on : https://www.cdc.gov/nchs/covid19/pulse/mental-health.htm

**Apply the scoring method on four feaures**

- Sum the score of 'INTEREST' and 'DOWN'. If the score is >= 3, the anticipant is associated with diagnoses of major depressive disorder. 
- Sum the score of 'ANXIOUS'  adn 'WORRY'. If the score is >=3, the anticipant is associated with diagnoses of generalized anxiety disorder.

#### 2.2.4 Drop missing and unknown values on these four critical features.
We don't want to see missing values and unknown values here in these four columns because they will not help us get the target columns, so we will need to drop them at the very beginning.

```python
#create a list
four_features = ['INTEREST','DOWN','ANXIOUS','WORRY']

#make a loop to drop each feature of these four and update to dataframe df2
for feature in four_features:
    mask = df2[feature].isin([-88, -99])
    df2 = df2[~mask]
```

#### 2.2.5 Get the target column of 'Major Depressive Disorder'

First, we will get the target column of Major Derpessive Disorder based on the feature 'INTEREST' and 'DOWN'. 

```python
# Sum the score of 'INTEREST' and 'DOWN'. 
df2['PHQ2_Score_Sum']=df2['INTEREST']+df2['DOWN']

# If the score is >= 3, the anticipant is associated with diagnoses of major depressive disorder.
df2['Major Depressive Disorder']=df2['PHQ2_Score_Sum'].map(lambda x:'Yes' if x>=3 else 'No')
```

However !! If we reference from the CDC report, we should know that even though these latino/spanish/hispanic people also consider themselves as white/black/asian/else, we should still consider them as latino/spanish/hispanic people at the first place. 

So, we are going to merge these two columns to create a new categorical column of Race/Hispanic ethnicity based on the categorical method from CDC report. Specifically, the logic behind the code is :
- if the value from `['RHISPANIC']` is 2, it will be transfered to 'Hispanic ethnicity' of the new column
- if the value from `['RHISPANIC']` is 1, the value of `['RRACE']` will be read and transfered into the related race categories of the new column. 


```python
# First, let's get the columns copied from df2
# Get column from df2
df3.loc[:, 'Hispanic'] = df2['RHISPANIC'].copy()
df3.loc[:, 'Race'] = df2['RRACE'].copy()
# Sanity Check
df3.head()
```

```python
#Replace back the numeric values with category names of Race column
# List the column values of Race

#1: White, Alone
#2: Black, Alone
#3: Asian, Alone
#4: Any other race alone, or race in combination


# Define the mapping dictionary
mapping = {
    1:'White', 
    2:'Black',
    3:'Asian',
    4:'Other'
          }

# Apply the mapping using a lambda function
df3['Race'] = df3['Race'].map(lambda x: mapping.get(x, x))


# Check the value counts
df3['Race'].value_counts()
```

```python
#Replace back the numeric values with category names of Hispanic Column
# List the column values of Race

#1:No, not of Hispanic, Latino, or Spanish origin  
#2:Yes, of Hispanic, Latino, or Spanish origin


# Define the mapping dictionary
mapping = {
    1:'Not Hispanic', 
    2:'Hispanic'
          }

# Apply the mapping using a lambda function
df3['Hispanic'] = df3['Hispanic'].map(lambda x: mapping.get(x, x))


# Check the value counts
df3['Hispanic'].value_counts()
```

```python
#Create new column to apply CDC category method
#first, apply two conditions to separate hispanic and non-hispanic people
values = np.where(df3['Hispanic'] == 'Not Hispanic', df3['Race'], np.where(df3['Hispanic'] == 'Hispanic', 'Hispanic', ''))

#second, assign the values to df3
df3['Race/Hispanic ethnicity'] = values

#sanity check
df3['Race/Hispanic ethnicity'].value_counts()
```

```python
#Replace back the numeric values with category names
# List the column values


#`'SHRTGMEDICAL1'`:
#1: Shortage of a medicine or medication that requires a prescription or is given by provider, pharmacist, or hospital 
#-99: Question seen but category not selected
#<.m>: Missing / Did not report"


# Define the mapping dictionary
mapping = {1: "Prescription Medicine Shortage", 
           -99:"Unknown",
           -88:"Unknown"
        }

# Apply the mapping using a lambda function
df3['SHRTGMEDICAL1'] = df3['SHRTGMEDICAL1'].map(lambda x: mapping.get(x, x))

#`'SHRTGMEDICAL2'`:
#1: Shortage of a medicine or medication that is sold over the counter (without a prescription)
#-99: Question seen but category not selected
#<.m>: Missing / Did not report"


# Define the mapping dictionary
mapping = {1: "Counter Medicine Shortage", 
           -99:"Unknown",
           -88:"Unknown"
        }

# Apply the mapping using a lambda function
df3['SHRTGMEDICAL2'] = df3['SHRTGMEDICAL2'].map(lambda x: mapping.get(x, x))

#`'SHRTGMEDICAL3'`:
#1: Shortage of a medical equipment or supplies used at home such as infusion pumps, glucose monitors, home ventilators, masks, gloves, etc 
#-99: Question seen but category not selected
#<.m>: Missing / Did not report"

# Define the mapping dictionary
mapping = {1: "Medical Equipment Shortage", 
           -99:"Unknown",
           -88:"Unknown"
        }

# Apply the mapping using a lambda function
df3['SHRTGMEDICAL3'] = df3['SHRTGMEDICAL3'].map(lambda x: mapping.get(x, x))


#`'SHRTGMEDICAL4'`:
#1: Shortage of other critical products, please specify 
#-99: Question seen but category not selected
#<.m>: Missing / Did not report"

# Define the mapping dictionary
mapping = {1: "Other Shortage", 
           -99:"Unknown",
           -88:"Unknown"
        }

# Apply the mapping using a lambda function
df3['SHRTGMEDICAL4'] = df3['SHRTGMEDICAL4'].map(lambda x: mapping.get(x, x))

#`'SHRTGMEDICAL5'`:
#1: None
#-99: Question seen but category not selected
#<.m>: Missing / Did not report"


# Define the mapping dictionary
mapping = {1: "None", 
           -99:"Unknown",
           -88:"Unknown"
        }

# Apply the mapping using a lambda function
df3['SHRTGMEDICAL5'] = df3['SHRTGMEDICAL5'].map(lambda x: mapping.get(x, x))

# Sanity Check
df3[['SHRTGMEDICAL1','SHRTGMEDICAL2','SHRTGMEDICAL3','SHRTGMEDICAL4','SHRTGMEDICAL5']].value_counts()
```
